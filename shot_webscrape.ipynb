```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import re
```

```python
driver = webdriver.Chrome(executable_path='/mnt/c/Windows/chromedriver.exe')
driver.get('https://www.basketball-reference.com/players/c/curryst01/shooting/2018')
soup = BeautifulSoup(driver.page_source, 'lxml')
# print(soup.prettify())
```

```python
shot_area = soup.find(class_='shot-area')
shots = shot_area.find_all('div')
# print(shot_area)
print(shots)
```

```python
def getShotMake(string):
    if(string == '●'):
        return True
    else:
        return False



# stuff to work off of
tip = shots[0]['tip'].split('<br>')
# format:
# [DATE_OPPONENT, QUARTER_TIME_REMAINING, DESCRIPTION, GAME_SCORE]
#
# example:
# ['Oct 17, 2017, GSW vs HOU', '1st Qtr, 8:09 remaining', 'Made 3-pointer from 25 ft', 'GSW now leads 15-7']


# make string
make_string = shots[0].get_text()
made_shot = getShotMake(make_string)


# date & opponent
date_vs = tip[0].split(', ')
date = datetime.strptime((date_vs[0]+ ", "+ date_vs[1]), '%b %d, %Y')
print(date)

match = re.search(r'(?<=vs\s)\w*', date_vs[2])
if(match):
    vs = match.group(0)
else:
    vs = None
print(vs)


# game time - quarter and time remaining in quarter
print(tip[1])
game_time = tip[1].split(', ')
q = re.search('\d', game_time[0])
if(q):
    quarter = q.group(0)
else:
    quarter = None

t = re.search('\d{,2}:\d{2}', game_time[1])
if(t):
    time_left = t.group(0)
else:
    time_left = None


info = {
    'date': date,
    'vs': vs,
    'shot_make': made_shot,
    'quarter': quarter,
    'time_left': time_left,
    'shot_description': tip[2],
    'game_score': tip[3],
}
print(info)
```
